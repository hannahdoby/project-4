```python
import pandas as pd
```

```python
movies = pd.read_csv(r"ml-25m (2)\ml-25m\movies.csv")
movies.head()
```

```python
genres = {}
```

```python
for index, row in movies.iterrows():
    genres[row["movieId"]] = row["genres"].split("|")
genres
```

```python
genre_list = ["Action",
"Adventure",
"Animation",
"Children's",
"Comedy",
"Crime",
"Documentary",
"Drama",
"Fantasy",
"Film-Noir",
"Horror",
"Musical",
"Mystery",
"Romance",
"Sci-Fi",
"Thriller",
"War",
"Western",
"(no genres listed)"]
```

```python
my_list = []
```

```python
for genre in genre_list:
    my_list = []
    for index in movies.index:
        movie = movies["movieId"][index]
        if genre in genres[movie]:
            my_list.append(1)
        else:
            my_list.append(0)
    movies[genre] = my_list
```

```python
movies = movies.drop(columns = ["genres", "title"])
movies.head()
```

```python
movies.to_csv(r"ml-25m (2)\ml-25m\movies_modified.csv")
```

```python
tag_names = pd.read_csv(r"ml-25m (2)/ml-25m/genome-tags.csv")
tag_list = tag_names["tagId"]
tag_scores = pd.read_csv(r"ml-25m (2)\ml-25m\genome-scores.csv")
tag_scores.head()
```

```python
tag_expansion = {}
for tag in tag_list:
    tag_expansion[tag]=[]
```

```python
movie_list = list(tag_scores["movieId"].unique())
len(movie_list)
```

```python
for movie in list(tag_scores["movieId"].unique())[1500:1800]:
    movie_score_list = tag_scores.loc[tag_scores["movieId"]==movie]["relevance"].values
    for tag in tag_list:
        movies.loc[movies["movieId"]==movie, tag] = movie_score_list[tag-1]

```
